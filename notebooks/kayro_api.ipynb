```python
import os
import math
import time
import requests
import feedparser
import yfinance as yf
import pandas as pd
import numpy as np

from urllib.parse import quote
from datetime import datetime, timedelta, timezone
from pytrends.request import TrendReq
from googleapiclient.discovery import build
from transformers import pipeline


ALPHA_VANTAGE_API_KEY = os.getenv("ALPHA_VANTAGE_API_KEY")
YOUTUBE_API_KEY = os.getenv("YOUTUBE_API_KEY")


finbert = pipeline(
    "sentiment-analysis",
    model="ProsusAI/finbert"
)


def clamp(value, minimum=-1.0, maximum=1.0):
    try:
        return max(min(float(value), maximum), minimum)
    except Exception:
        return 0.0


def normalize_0_100(score_minus_1_to_1):
    return round(((clamp(score_minus_1_to_1) + 1) / 2) * 100, 2)


def finbert_score(text: str):
    if not text or not isinstance(text, str) or not text.strip():
        return 0.0, "neutral", 0.0

    result = finbert(text[:512])[0]

    label = result["label"].lower()
    confidence = float(result["score"])

    if label == "positive":
        score = confidence
    elif label == "negative":
        score = -confidence
    else:
        score = 0.0

    return score, label, confidence


def average(values):
    values = [v for v in values if v is not None]
    return float(np.mean(values)) if values else 0.0


# =========================
# 1. GOOGLE NEWS RSS
# =========================

def fetch_google_news(query: str, limit: int = 20):
    rss_url = f"https://news.google.com/rss/search?q={quote(query)}&hl=en-US&gl=US&ceid=US:en"
    feed = feedparser.parse(rss_url)

    articles = []

    for item in feed.entries[:limit]:
        articles.append({
            "source": "google_news",
            "title": item.get("title", ""),
            "link": item.get("link", ""),
            "published": item.get("published", "")
        })

    return articles


def score_google_news(query: str, limit: int = 20):
    articles = fetch_google_news(query, limit)

    scores = []

    for article in articles:
        score, label, confidence = finbert_score(article["title"])
        article["sentiment_score"] = score
        article["sentiment_label"] = label
        article["confidence"] = confidence
        scores.append(score)

    return {
        "score": average(scores),
        "article_count": len(articles),
        "articles": articles[:5]
    }


# =========================
# 2. YAHOO FINANCE NEWS
# =========================

def fetch_yahoo_news(ticker: str, limit: int = 20):
    try:
        news = yf.Ticker(ticker).news or []
    except Exception:
        news = []

    articles = []

    for item in news[:limit]:
        articles.append({
            "source": "yahoo_finance",
            "title": item.get("title", ""),
            "publisher": item.get("publisher", ""),
            "link": item.get("link", ""),
            "published": item.get("providerPublishTime")
        })

    return articles


def score_yahoo_news(ticker: str, limit: int = 20):
    articles = fetch_yahoo_news(ticker, limit)

    scores = []

    for article in articles:
        score, label, confidence = finbert_score(article["title"])
        article["sentiment_score"] = score
        article["sentiment_label"] = label
        article["confidence"] = confidence
        scores.append(score)

    return {
        "score": average(scores),
        "article_count": len(articles),
        "articles": articles[:5]
    }


# =========================
# 3. ALPHA VANTAGE NEWS SENTIMENT
# =========================

def score_alpha_vantage(ticker: str, limit: int = 50):
    if not ALPHA_VANTAGE_API_KEY:
        return {
            "score": 0.0,
            "article_count": 0,
            "articles": [],
            "error": "Missing ALPHA_VANTAGE_API_KEY"
        }

    url = "https://www.alphavantage.co/query"

    params = {
        "function": "NEWS_SENTIMENT",
        "tickers": ticker,
        "apikey": ALPHA_VANTAGE_API_KEY
    }

    try:
        response = requests.get(url, params=params, timeout=20)
        data = response.json()
    except Exception as error:
        return {
            "score": 0.0,
            "article_count": 0,
            "articles": [],
            "error": str(error)
        }

    feed = data.get("feed", [])[:limit]

    scores = []
    articles = []

    for item in feed:
        raw_score = item.get("overall_sentiment_score", 0)

        try:
            score = float(raw_score)
        except Exception:
            score = 0.0

        scores.append(score)

        articles.append({
            "source": "alpha_vantage",
            "title": item.get("title", ""),
            "url": item.get("url", ""),
            "published": item.get("time_published", ""),
            "sentiment_score": score,
            "sentiment_label": item.get("overall_sentiment_label", "")
        })

    return {
        "score": clamp(average(scores)),
        "article_count": len(feed),
        "articles": articles[:5]
    }


# =========================
# 4. GOOGLE TRENDS
# =========================

def score_google_trends(keyword: str):
    try:
        pytrends = TrendReq(hl="en-US", tz=360)

        pytrends.build_payload(
            [keyword],
            timeframe="today 3-m",
            geo="US"
        )

        df = pytrends.interest_over_time()

        if df.empty or keyword not in df.columns:
            return {
                "score": 0.0,
                "interest": 0,
                "average": 0,
                "error": None
            }

        values = df[keyword].dropna()

        if len(values) < 10:
            return {
                "score": 0.0,
                "interest": 0,
                "average": 0,
                "error": None
            }

        current = float(values.iloc[-1])
        avg = float(values.mean())
        recent_avg = float(values.tail(14).mean())

        trend_vs_average = (current - avg) / max(avg, 1)
        recent_momentum = (recent_avg - avg) / max(avg, 1)

        score = clamp(
            0.6 * trend_vs_average +
            0.4 * recent_momentum
        )

        return {
            "score": score,
            "interest": current,
            "average": avg,
            "recent_average": recent_avg,
            "error": None
        }

    except Exception as error:
        return {
            "score": 0.0,
            "interest": 0,
            "average": 0,
            "error": str(error)
        }


# =========================
# 5. YOUTUBE
# =========================

def fetch_youtube_videos(query: str, limit: int = 10):
    if not YOUTUBE_API_KEY:
        return []

    try:
        youtube = build(
            "youtube",
            "v3",
            developerKey=YOUTUBE_API_KEY
        )

        request = youtube.search().list(
            q=query,
            part="snippet",
            type="video",
            maxResults=limit,
            order="date"
        )

        response = request.execute()

    except Exception:
        return []

    videos = []

    for item in response.get("items", []):
        snippet = item.get("snippet", {})

        videos.append({
            "source": "youtube",
            "title": snippet.get("title", ""),
            "description": snippet.get("description", ""),
            "channel": snippet.get("channelTitle", ""),
            "published": snippet.get("publishedAt", "")
        })

    return videos


def score_youtube(query: str, limit: int = 10):
    videos = fetch_youtube_videos(query, limit)

    scores = []

    for video in videos:
        text = f"{video['title']}. {video['description']}"
        score, label, confidence = finbert_score(text)

        video["sentiment_score"] = score
        video["sentiment_label"] = label
        video["confidence"] = confidence

        scores.append(score)

    return {
        "score": average(scores),
        "video_count": len(videos),
        "videos": videos[:5],
        "error": None if YOUTUBE_API_KEY else "Missing YOUTUBE_API_KEY"
    }


# =========================
# 6. WIKIPEDIA PAGEVIEWS
# =========================

def fetch_wikipedia_pageviews(page_title: str, days: int = 30):
    end = datetime.now(timezone.utc) - timedelta(days=1)
    start = end - timedelta(days=days)

    start_str = start.strftime("%Y%m%d")
    end_str = end.strftime("%Y%m%d")

    title = quote(page_title.replace(" ", "_"), safe="")

    url = (
        "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/"
        f"en.wikipedia/all-access/user/{title}/daily/{start_str}/{end_str}"
    )

    headers = {
        "User-Agent": "KayroStockSentiment/1.0"
    }

    try:
        response = requests.get(url, headers=headers, timeout=20)

        if response.status_code != 200:
            return []

        data = response.json()

    except Exception:
        return []

    return [
        item.get("views", 0)
        for item in data.get("items", [])
    ]


def score_wikipedia(page_title: str):
    views = fetch_wikipedia_pageviews(page_title, days=60)

    if len(views) < 20:
        return {
            "score": 0.0,
            "views_latest": 0,
            "views_average": 0,
            "views_total": sum(views)
        }

    latest_7d = np.mean(views[-7:])
    previous_avg = np.mean(views[:-7])

    score = clamp(
        (latest_7d - previous_avg) / max(previous_avg, 1)
    )

    return {
        "score": score,
        "views_latest": float(latest_7d),
        "views_average": float(previous_avg),
        "views_total": int(sum(views))
    }


# =========================
# 7. GOOGLE DISCOVER / MEDIA BUZZ PROXY
# =========================

def score_google_discover_proxy(query: str):
    today_articles = fetch_google_news(
        query=query,
        limit=50
    )

    article_count = len(today_articles)

    buzz_score = min(article_count / 50, 1.0)

    sentiment_scores = []

    for article in today_articles:
        score, _, _ = finbert_score(article["title"])
        sentiment_scores.append(score)

    sentiment = average(sentiment_scores)

    combined = clamp(
        0.5 * sentiment +
        0.5 * buzz_score
    )

    return {
        "score": combined,
        "buzz_score": buzz_score,
        "sentiment_score": sentiment,
        "article_count": article_count
    }


# =========================
# FINAL KAYRO SOCIAL SCORE
# =========================

def build_kayro_social_score(
    ticker: str,
    company_name: str
):
    ticker = ticker.upper()

    google_news = score_google_news(
        query=f"{ticker} stock OR {company_name} stock",
        limit=25
    )

    yahoo_news = score_yahoo_news(
        ticker=ticker,
        limit=20
    )

    alpha_vantage = score_alpha_vantage(
        ticker=ticker,
        limit=50
    )

    google_trends = score_google_trends(
        keyword=company_name
    )

    youtube = score_youtube(
        query=f"{company_name} stock",
        limit=10
    )

    wikipedia = score_wikipedia(
        page_title=company_name
    )

    google_discover = score_google_discover_proxy(
        query=f"{company_name} stock"
    )

    final_score_minus_1_to_1 = clamp(
        0.22 * google_news["score"] +
        0.18 * yahoo_news["score"] +
        0.20 * alpha_vantage["score"] +
        0.12 * google_trends["score"] +
        0.10 * youtube["score"] +
        0.08 * wikipedia["score"] +
        0.10 * google_discover["score"]
    )

    return {
        "ticker": ticker,
        "company_name": company_name,
        "social_score": round(final_score_minus_1_to_1, 4),
        "social_score_100": normalize_0_100(final_score_minus_1_to_1),
        "components": {
            "google_news": {
                "score": round(google_news["score"], 4),
                "article_count": google_news["article_count"]
            },
            "yahoo_finance": {
                "score": round(yahoo_news["score"], 4),
                "article_count": yahoo_news["article_count"]
            },
            "alpha_vantage": {
                "score": round(alpha_vantage["score"], 4),
                "article_count": alpha_vantage["article_count"]
            },
            "google_trends": {
                "score": round(google_trends["score"], 4),
                "interest": google_trends["interest"],
                "error": google_trends.get("error")
            },
            "youtube": {
                "score": round(youtube["score"], 4),
                "video_count": youtube["video_count"],
                "error": youtube.get("error")
            },
            "wikipedia": {
                "score": round(wikipedia["score"], 4),
                "views_total": wikipedia["views_total"]
            },
            "google_discover_proxy": {
                "score": round(google_discover["score"], 4),
                "article_count": google_discover["article_count"]
            }
        },
        "top_items": {
            "google_news": google_news["articles"],
            "yahoo_finance": yahoo_news["articles"],
            "alpha_vantage": alpha_vantage["articles"],
            "youtube": youtube["videos"]
        }
    }
```
